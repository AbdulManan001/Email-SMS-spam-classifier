## Data Preprocessing
- Lower case
- Tokenization
- Removing special Characters
- Removing Stop words(is, the, etc) and punctuation
- Stemming(remove same type of words e.g dancing, danced, dance ,
all three are converted into a basic word i.e danc)

```python
# Downloading Stopwords Module
import nltk
nltk.download('stopwords')
# Importing Stemming Module
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
from nltk.corpus import stopwords 
import string # for Punctuaton

def tranform_text(text):
    #Lower Case
    text = text.lower()

    #Tokenization
    text = nltk.word_tokenize(text)
    
    #Removing Special Characters
    y = []
    for i in text:
        if i.isalnum():
            y.append(i)
    text = y[:]
    y.clear()

    # Removing Stopwords and Punctuation
    for i in text:
        if i not in stopwords.words('english') and i not in string.punctuation:
            y.append(i)
    text = y[:]
    y.clear()

    #Stemming
    for i in text:
        y.append(ps.stem(i))

    return " ".join(y)
```

```python
tranform_text(df['text'][10])
```

```python
df['text'][10]
```

```python
df['transformed_text'] = df['text'].apply(tranform_text)
```

## 4. Model Building

```python
%pip install xgboost
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from xgboost import XGBClassifier
```

```python
# Vectorization
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
X = cv.fit_transform(df['transformed_text']).toarray()
```

```python

y = df['target']
```

```python
from sklearn.model_selection import train_test_split
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
dtc = DecisionTreeClassifier(max_depth = 5)
rfc = RandomForestClassifier(n_estimators=100, random_state=2)
lg = LogisticRegression()
xgb = XGBClassifier(n_estimators= 100, random_state= 2)
```

```python
from sklearn.metrics import precision_score
from sklearn.metrics import accuracy_score


def train_classifier(clf, X_train, y_train, X_test, y_test):
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    print(accuracy_score(y_test, y_pred))
    print(precision_score(y_test, y_pred))
    return y_pred
```

```python
# Decision Tree Algorithm
train_classifier(dtc, X_train, y_train, X_test, y_test)
```

```python
# Logistic Regression
train_classifier(lg, X_train, y_train, X_test, y_test)
```

```python
# XGBoost Algorithm
train_classifier(xgb, X_train, y_train, X_test, y_test)
```

```python
# Random Forest Algorithm
train_classifier(rfc, X_train, y_train, X_test, y_test)
```
